```python
import pandas as pd
import numpy as np
from os import path

data_path = '../features/2019-04-04_202426'
feature_path = 'features.pkl'
label_path = 'labels.npy'

features = pd.read_pickle(path.join(data_path, feature_path))
labels = np.load(path.join(data_path, label_path))
```

```python
import classification
from IPython.display import display

clfs = classification.Classifiers(features,labels,[])

```

## Observation:
The sklearn mutual_information_gain function function introduces some kind of randomness: adding small noise to continuous variables in order to remove repeated values. 
Thus to have reliable results, we need to repeat this a few times

**IMPORTANT NOTE:** info gain can differ in results as this is a randomised process. Order should be the same!!



```python
info_gain = clfs.repeat_info_gain(features, repeats=20)
```

```python
pd.options.display.max_rows=1000
display(info_gain.loc[:, ['Feature Name', 'Mean']])
```
